```python
import numpy as np
from nilearn import datasets
from nilearn.input_data import NiftiLabelsMasker
from nilearn.image import index_img
from sklearn.preprocessing import StandardScaler
import nibabel as nib
from matplotlib import pyplot as plt
import sklearn

brain_data = datasets.fetch_oasis_vbm(n_subjects=100)
crad = datasets.fetch_atlas_craddock_2012()
output_variable = StandardScaler().fit_transform(brain_data.ext_vars.age[:, None])[:,0] # gives subject age on standard units after z-scoring

data_n = output_variable.size
sets_n = 43

input_variables = [None]*sets_n

for i in range(sets_n):

    atlas_nii = index_img(crad['scorr_mean'], i); #this needs to be modified for some tasks
    masker = NiftiLabelsMasker(labels_img= atlas_nii, standardize=True);
    input_variables[i] = masker.fit_transform(brain_data.gray_matter_maps);
    print('%i regions found in this atlas' % (len(np.unique(atlas_nii.get_fdata().ravel())) - 1))
```

```python
import sklearn
from sklearn import cross_decomposition
from sklearn import discriminant_analysis
from sklearn import pipeline

trainers = [None]*3
trainers[0] = sklearn.cross_decomposition.PLSRegression(n_components=3)
trainers[1] = sklearn.pipeline.make_pipeline(sklearn.preprocessing.StandardScaler(),sklearn.decomposition.PCA(n_components=3),sklearn.linear_model.LinearRegression());
trainers[2] = sklearn.cross_decomposition.CCA(n_components=1)

names = ['PLSR','PCR','CCA']
```

n_components=1 for CCa because we are trying to predict sexe which is 1 dimention.

Get following error otherwise : 'n_components' upper bound is 1. Got 2 instead. Reduce 'n_components'

```python
#task 6

out = output_variable

insamp_accuracies = np.zeros((3,sets_n))

for i,inp in enumerate(input_variables):
    for j in range(3):
        model = trainers[j].fit(inp,out)

        prediction = model.predict(inp).flatten()

        insamp_accuracies[j,i] = sklearn.metrics.r2_score(out, prediction)
```

```python
plt.plot(insamp_accuracies.transpose());
plt.legend(names)
plt.xlabel('Data set')
plt.ylabel('R^2 value')
plt.title('Model accuracy by dataset');
```

```python
#task 7

out = output_variable

EIGHTY = int(out.size*0.8)
linspace = np.arange(out.size)
train = linspace < EIGHTY
test = linspace >= EIGHTY

outsamp_accuracies = np.zeros((3,sets_n))
insamp_accuracies = np.zeros((3,sets_n))


splitter = sklearn.model_selection.KFold(n_splits=5, shuffle=False)

for i,inp in enumerate(input_variables):

    index_generator = splitter.split(inp,out)

    for (train,test) in index_generator:

        for j in range(3):

            model = trainers[j].fit(inp[train],out[train])

            prediction = model.predict(inp[test]).flatten()
            outsamp_accuracies[j,i] += sklearn.metrics.r2_score(out[test], prediction)

            prediction = model.predict(inp[train]).flatten()
            insamp_accuracies[j,i] += sklearn.metrics.r2_score(out[train], prediction)

outsamp_accuracies*=0.2
insamp_accuracies*=0.2
```

```python
plt.plot(insamp_accuracies.transpose());
plt.plot(outsamp_accuracies.transpose());
plt.legend(['in sample PLSR','in sample PCR','in sample CCA','out sample PLSR','out sample PCR','out sample CCA'])

plt.xlabel('Data set')

plt.ylabel('R^2 value')
plt.title('Model accuracy by dataset');
```

```python
#task 8

out = output_variable

EIGHTY = int(out.size*0.8)
linspace = np.arange(out.size)
train = linspace < EIGHTY
test = linspace >= EIGHTY

outsamp_accuracies = np.zeros((3,sets_n))
insamp_accuracies = np.zeros((3,sets_n))


splitter = sklearn.model_selection.KFold(n_splits=5, shuffle=False)
pca_trainer = sklearn.decomposition.PCA(n_components=3)

for i,inp in enumerate(input_variables):

    index_generator = splitter.split(inp,out)

    for (train,test) in index_generator:

        for j in range(3):
            
            pca = pca_trainer.fit(inp[train])
            inp_train = pca.transform(inp[train])
            inp_test = pca.transform(inp[test])

            model = trainers[j].fit(inp_train,out[train])

            prediction = model.predict(inp_test).flatten()
            outsamp_accuracies[j,i] += sklearn.metrics.r2_score(out[test], prediction)

            prediction = model.predict(inp_train).flatten()
            insamp_accuracies[j,i] += sklearn.metrics.r2_score(out[train], prediction)

outsamp_accuracies*=0.2
insamp_accuracies*=0.2
```

```python
plt.plot((insamp_accuracies).transpose() +np.array([-0.005,+0.005,0]) );
plt.plot(outsamp_accuracies.transpose() +np.array([-0.005,+0.005,0]) );
plt.legend(['in sample PLSR','in sample PCR','in sample CCA','out sample PLSR','out sample PCR','out sample CCA'])

plt.xlabel('Data set')

plt.ylabel('R^2 value')
plt.title('Model accuracy by dataset using PCA');
```

```python
#task 9

out = output_variable

EIGHTY = int(out.size*0.8)
linspace = np.arange(out.size)
train = linspace < EIGHTY
test = linspace >= EIGHTY

splitter = sklearn.model_selection.KFold(n_splits=5, shuffle=False)

PLSR_coefs = 0
PCR_coefs = 0
CCA_coefs = 0


inp = input_variables[0]

index_generator = splitter.split(inp,out)

for (train,test) in index_generator:

    PLSR_model = trainers[0].fit(inp[train],out[train])
    PLSR_coefs += PLSR_model.x_loadings_/5

    PCR_model = trainers[1].fit(inp[train],out[train])
    PCR_coefs += PCR_model.named_steps['linearregression'].coef_/5

    CCA_model = trainers[2].fit(inp[train],out[train])
    CCA_coefs += CCA_model.coef_/5
```

```python
plt.imshow(PLSR_coefs, interpolation='nearest', aspect='auto');
plt.xticks([])
plt.yticks([])
plt.title('PLSR coeficients')
plt.xlabel('Components')
plt.ylabel('Regions')
plt.colorbar()
plt.show()
plt.imshow(PCR_coefs.reshape(1,3), interpolation='nearest', aspect='auto');
plt.xticks([])
plt.yticks([])
plt.title('PCR coeficients')
plt.xlabel('Components')
plt.ylabel('Regions')
plt.colorbar()
plt.show()
plt.imshow(CCA_coefs.reshape(10,1), interpolation='nearest', aspect='auto');
plt.xticks([])
plt.yticks([])
plt.title('CCA coeficients')
plt.xlabel('Components')
plt.ylabel('Regions')
plt.colorbar()
plt.show()
```
